# Import Libraries

```python
import pandas as pd
```

# Extract Source Data

```python
# Extract data from a all Excel files
business_services_vertical_df = pd.read_excel("/Users/ankur/Documents/GitHub/data-migration-assessment/data/source/Business Services Pipeline.xlsx", skiprows=5)
consumer_retail_and_healthcare_vertical_df = pd.read_excel("/Users/ankur/Documents/GitHub/data-migration-assessment/data/source/Consumer Retail and Healthcare Pipeline.xlsx",skiprows=8).drop(columns=['Unnamed: 0'])
contacts_df1 = pd.read_excel("/Users/ankur/Documents/GitHub/data-migration-assessment/data/source/Contacts.xlsx", sheet_name="Tier 1's")
contacts_df2 = pd.read_excel("/Users/ankur/Documents/GitHub/data-migration-assessment/data/source/Contacts.xlsx", sheet_name="Tier 2's")
events_df1 = pd.read_excel("/Users/ankur/Documents/GitHub/data-migration-assessment/data/source/Events.xlsx",sheet_name="Leaders and Partners Dinner")
events_df2 = pd.read_excel("/Users/ankur/Documents/GitHub/data-migration-assessment/data/source/Events.xlsx",sheet_name="2019 Market Re-Cap")
private_equity_companies_df = pd.read_excel("/Users/ankur/Documents/GitHub/data-migration-assessment/data/source/PE Comps.xlsx",skiprows=2)
```

# Initial Data Cleansing

```python
# Cleansing the business services vertical DataFrame to add a new column
business_services_vertical_df['VerticalDepartment'] = "Business Services"
bs_vertical_cleansed_df = business_services_vertical_df

# Removed NULL rows from the consumer retail and healthcare vertical DataFrame
consumer_retail_and_healthcare_vertical_df = consumer_retail_and_healthcare_vertical_df[consumer_retail_and_healthcare_vertical_df['Company Name'].notna() & consumer_retail_and_healthcare_vertical_df['Lead MD'].notna()]

#Add a new column to the cleansed DataFrame
consumer_retail_and_healthcare_vertical_df['VerticalDepartment'] = "Consumer Retail and Healthcare"
crh_vertical_cleansed_df = consumer_retail_and_healthcare_vertical_df

# Concatenate the contacts DataFrames
contacts_cleansed_df = pd.concat([contacts_df1, contacts_df2], ignore_index=True)

# Set the 'EventType' for each event DataFrame
events_df1['EventType'] = 'Leaders and Partners Dinner'
events_df2['EventType'] = '2019 Market Re-Cap'

# Concatenate the events DataFrames
events_cleansed_df = pd.concat([events_df1, events_df2], ignore_index=True)

# Cleansing the private equity companies DataFrame
private_equity_companies_cleansed_df = private_equity_companies_df.copy()
```

```python
# Concatenate the cleansed vertical DataFrames
all_vertical_deals_df = pd.concat([bs_vertical_cleansed_df, crh_vertical_cleansed_df], ignore_index=True)
```
